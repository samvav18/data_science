```python
data.isnull().sum() # it gives the number of null values
```

```python
# pass the columns which have null values in a list
null_list=[]
for col in data.columns:
    if data[col].isnull().sum()>0:
        null_list.append(col)
print(null_list)
    
```

```python
#calculate the null value percentage
null_dict={}
for col in data.columns:
    missing_count=data[col].isnull().sum()
    if missing_count > 0:
        missing_per=(missing_count/len(data))*100
        null_dict[col]=round(missing_per,3)

for col,per in null_dict.items():
    print(f"{col}:{per}%")
```
